```python
import numpy as np
import pandas as pd

df = pd.read_excel('Cleaning.xlsx')
print(df.columns)
# print(df.shape)
# print(df.index)

dropna_all = df.dropna(how='all')
dropna_all

# dropna_any = df.dropna(how='any')
# dropna_any

# ->>>>>>>> using this to put a place that null in .id_fillna() put 0
df['id'] = df['id'].fillna(0).astype(int)

# ---->>>>>> using to make a name to the first letter is upper
df['customer name'] = df['customer name'].str.strip().str.title()

# --->>>>>> using this to break an email 2 part
df[['User', 'Domain']] = df['email'].str.split('@', expand=True)
# print(df[['User', 'Domain']])

## ---->>>>>>>>>> Because our data have only 1100 and the null and worng standard is 910 ,so i decision to fill in by uisng {.ffill } 
df['purchase date'] = pd.to_datetime(df['purchase date'], errors='coerce')
df['purchase date'] = df['purchase date'].ffill()

# using to set the ['purchase date' ] as month/day/year
df['purchase date'] = df['purchase date'].dt.strftime('%m/%d/%Y')
df.dropna(subset=['purchase date'])         # if null delete 

# -->>>>>>> using to covert price to double and delete '$'
df['price'] = df['price'].str.replace('$','', regex=False).str.replace(',','',regex=False)
df['price'] = pd.to_numeric(df['price'])

# -->>>>>> using to fill the Nat place of 'trx'
df['trx'] = df['trx'].ffill()

# --->>>>>>> uisng to fill shipping state
df['shipping state'] = df['shipping state'].ffill()

df['User'] = df['User'].ffill()

df['Domain'] = df['Domain'].ffill()

#      --->>>>>>>>>   using to see duplicated in the row 
duplicate_rows = df[df.duplicated()]
# duplicate_rows.shape 
# duplicate_rows.index

# after we see the duplicated in the columns already i want to filter it out it mean take it out 
data_clean = df.drop_duplicates(keep='first')
data_clean

data_clean.to_excel('AfterFiltering.xlsx', index=False)

# using to count the columns that NaT
# print(df[['trx']])
# Nat_count = df['User'].isnull().sum()
# print(nat_count)

# so trx is null 72 and shipping state is null 244
```
