# **Data Cleaning**

## Detecting Missing Values

Before you can fix missing data, you need to find it. These methods return boolean values (True or False).
- .isna() and .isnull(): These do exactly the same thing. They return True if the data is missing (NaN), and False if it's not.
- .notna(): The exact opposite. It returns True if the data is present, and False if it is missing.

```python
import pandas as pd
import numpy as np

# Creating a sample dataset with missing values (NaNs)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', np.nan],
    'Age': [25, np.nan, 30, np.nan],
    'Score': [85, 90, np.nan, np.nan]
}
df = pd.DataFrame(data)
```

## Dropping Missing Values (.dropna())

Sometimes, the best approach is to simply delete the rows or columns that have missing data.
- how='any' (Default): Drops the row if even one value is missing. Use this when you need perfectly complete records.
- how='all': Drops the row only if every single value in that row is missing. Use this to clean up completely blank rows.
- subset: Only looks at specific columns to decide if a row should be dropped.

```python
import pandas as pd
import numpy as np

# Creating a sample dataset with missing values (NaNs)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', np.nan],
    'Age': [25, np.nan, 30, np.nan],
    'Score': [85, 90, np.nan, np.nan]
}
df = pd.DataFrame(data)
```

## Filling Missing Values (.fillna())

If you don't want to lose data by dropping rows, you can fill the blanks with substitute values.
- Constants: Replace NaNs with a specific number or string (like 0 or "Unknown").
- Mean/Median: Replace NaNs with the mathematical average or middle value of that column. Great for numeric data so you don't skew your statistics.
- Forward Fill (ffill): Takes the last known value and copies it forward into the blank space.
- Backward Fill (bfill): Takes the next known value and copies it backward into the blank space.

```python
import pandas as pd
import numpy as np

# Creating a sample dataset with missing values (NaNs)
data = {
    'Name': ['Alice', np.nan, 'Charlie', 'Bob'],
    'Age': [25, np.nan, 30, np.nan],
    'Score': [85, 90, np.nan, np.nan]
}
df = pd.DataFrame(data)
```

## Identifying Duplicates: .duplicated()

It scans your data row by row. It returns True if a row is an exact copy of a row that appeared earlier, and False if it is unique.

```python
import pandas as pd

# Our sample data contains duplicate rows for 'Alice' and 'Bob'
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Age': [25, 30, 25, 35, 30]
}
df = pd.DataFrame(data)
```

## Removing Duplicates: .drop_duplicates()

It returns a new version of your dataset with the duplicate rows completely removed.
The keep parameter: This tells Pandas which version of the duplicate you want to save.
- keep='first' (Default): Keeps the very first time it sees the row. Drops the rest.
- keep='last': Keeps the very last time it sees the row. Drops the earlier ones.
- keep=False: Drops all of them. If a row isn't 100% unique, it gets completely deleted.

```python
import pandas as pd

# Our sample data contains duplicate rows for 'Alice' and 'Bob'
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Age': [25, 30, 25, 35, 30]
}
df = pd.DataFrame(data)
```

## .astype()

You use it when your data is relatively clean, but it's just in the wrong format. It is highly effective for saving memory (like converting a column of 100 unique text labels into a category type).

```python
import pandas as pd

# Sample data: Ages are floats (e.g., 25.0)
df = pd.DataFrame({'Age': [25.0, 30.0, 22.0]})
```

## pd.to_numeric(errors='coerce')

This is your "safe" number converter for messy data. If Pandas tries to convert a column to numbers and hits a random word like "Unknown" or "Apple", the code normally crashes. By adding errors='coerce', you are telling Pandas: "If you can't turn it into a number, just replace it with a blank (NaN) and keep going."

```python
df = pd.DataFrame({'Price': ['15', '20', 'Free', '35']})
```

## pd.to_datetime()

This turns "dumb" text dates into "smart" date objects. If your date is just a string ("2023-10-15"), Python just sees it as a word. Once you use to_datetime(), Python recognizes it as a real point in time, allowing you to easily pull out just the month, the year, or even calculate the days between two dates.

```python
df = pd.DataFrame({'Signup_Date': ['2023-01-15', '2024-05-20']})
```

## Basic Text Formatting

These methods are perfect for cleaning up messy, inconsistent text data.
- .str.lower(): Converts all characters in the column to lowercase.
- .str.upper(): Converts all characters in the column to uppercase.
- .str.strip(): Removes any accidental white spaces from the very beginning or end of the text.

```python
import pandas as pd

# Creating a messy column of text
df = pd.DataFrame({'fruits': ['  Apple ', 'bAnAnA', ' CHERRY  ']})
```

## Splitting and Replacing

These methods help you extract specific information or swap out unwanted characters.
- .str.split('x'): Chops the string into a list of smaller strings wherever the character 'x' appears (like a space, comma, or dash). If you don't put anything in the brackets, it splits by spaces.
- .str.replace('old', 'new'): Swaps out a specific character or pattern.

**Note on Regex**: Regular Expressions (Regex) are special codes used to find complex patterns. For example, \d+ is regex for "any numbers". You can use .str.replace(r'\d+', '') to easily erase all numbers from a string!

```python
import pandas as pd

# Creating a column with combined info
df = pd.DataFrame({'user_info': ['John Doe', 'Jane Smith', 'Bob99']})
```

# **Data Manipulation & Transformation**

## Adding New Columns

You can create a brand new column by doing math or combining data from the columns you already have. You simply declare the new column name in brackets and set it equal to your operation.

```python
import pandas as pd

data = {'Price': [10, 20, 30], 'Tax': [1, 2, 3]}
df = pd.DataFrame(data)
```

## Renaming Columns: .rename()

This allows you to change the names of your columns. You pass a dictionary to the columns parameter where the key is the old name and the value is the new name.

**Note**: Using inplace=True modifies your original DataFrame directly. If you don't use it, pandas creates a copy instead.

```python
import pandas as pd

data = {'Price': [10, 20, 30], 'Tax': [1, 2, 3]}
df = pd.DataFrame(data)
```

## Dropping Rows or Columns: .drop()

This is how you delete data. The key here is the axis parameter:
- axis=1 means you are dropping a column.
- axis=0 means you are dropping a row (based on its index number).

```python
import pandas as pd

data = {'Price': [10, 20, 30], 'Tax': [1, 2, 3]}
df = pd.DataFrame(data)
```

## Using .apply()

A versatile tool that takes a function and applies it to an entire column (Series), or across the rows/columns of a whole DataFrame. Think of it as a for loop, but much faster and written in a single line.

**Best for**: Complex, custom logic that needs to evaluate data row-by-row or column-by-column.

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Dept_Code': ['HR', 'IT', 'IT']
})
```

## Using .map()

A method used only on a single column (Series). It looks at every value in that column and replaces it with a new value based on a "key" you provide (usually a Python dictionary).

**Best for**: Direct translations, standardizing categories, or substituting values.

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Dept_Code': ['HR', 'IT', 'IT']
})
```

## Using lambda functions

A "throwaway," one-line function that doesn't need a name. Instead of writing a full def my_function(): block, you write the logic right where you need it.

**Best for**: Quick, simple transformations (like basic math or text formatting) used directly inside .apply() or .map().

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Dept_Code': ['HR', 'IT', 'IT']
})
```

## The .groupby() Object (The "Split")

It splits your DataFrame into separate "buckets" based on the unique values in a specific column. By itself, it doesn't calculate anything; it just preps the data.

```python
import pandas as pd

data = {
    'Team': ['A', 'A', 'B', 'B'],
    'Player': ['John', 'Jane', 'Bob', 'Alice'],
    'Points': [10, 20, 15, 25]
}
df = pd.DataFrame(data)
```

## Aggregating with .agg() (The "Combine into Summary")

It takes your grouped buckets, applies calculations to them, and shrinks the data down to a single summary row per group. Use this when you want to know the "bottom line" for each category.

```python
import pandas as pd

data = {
    'Team': ['A', 'A', 'B', 'B'],
    'Player': ['John', 'Jane', 'Bob', 'Alice'],
    'Points': [10, 20, 15, 25]
}
df = pd.DataFrame(data)
```

## Transforming with .transform() (The "Combine into Original Shape")

It performs a calculation on each group, but instead of shrinking the data down, it returns a value for every single original row. Use this when you want to compare individual rows to their group's overall performance without losing your original dataset.

```python
import pandas as pd

data = {
    'Team': ['A', 'A', 'B', 'B'],
    'Player': ['John', 'Jane', 'Bob', 'Alice'],
    'Points': [10, 20, 15, 25]
}
df = pd.DataFrame(data)
```

## Concatenation: pd.concat()

Think of this as physical glue. You are taking two separate DataFrames and simply stacking them on top of each other (vertically) or placing them side-by-side (horizontally). It doesn't care about matching specific values; it just sticks the blocks together.

**Best used when**: You have identical columns in two files (like "January Sales" and "February Sales") and just want to combine them into one long list.

```python
import pandas as pd

# Two separate dataframes with the same columns
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df2 = pd.DataFrame({'Name': ['Charlie', 'David'], 'Age': [35, 40]})
```

## Merging/Joining: pd.merge()

Think of this like a zipper or an Excel VLOOKUP. You use pd.merge() when you have two datasets that share a common column (a "key"), and you want to align the data based on those matching values.
- Inner: Keep only the rows where the key exists in both tables.
- Outer: Keep all rows from both tables, filling in missing data with NaN.
- Left/Right: Keep all rows from the Left (or Right) table, and only pull in matching data from the other.

**Best used when**: You have related information in different tables (e.g., one table with Employee IDs and Names, and another table with Employee IDs and Salaries).

```python
import pandas as pd

# Table 1: Employees
employees = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})

# Table 2: Salaries
salaries = pd.DataFrame({'ID': [1, 2, 4], 'Salary': [50000, 60000, 70000]})
```

## Pivot Tables (pd.pivot_table())

Think of a pivot table as a way to summarize and reorganize a messy, detailed dataset into a clean summary table. You tell pandas what you want as your rows (index), what you want as your columns (columns), and how you want to calculate the data in the middle (values and aggfunc, like sum or average). It takes "long" data and spreads it out "wide."

```python
import pandas as pd

# 1. Our starting "long" data
df = pd.DataFrame({
    'Region': ['North', 'North', 'South', 'South', 'North'],
    'Fruit': ['Apple', 'Banana', 'Apple', 'Banana', 'Apple'],
    'Sales': [10, 15, 20, 25, 5]
})
```

## Melt (pd.melt())

Melt is the exact opposite of a pivot table. It takes "wide" data (where information is spread across many columns) and squashes it down into "long" data. You keep an identifier column untouched (id_vars), and the rest of the columns get melted down into two new columns: one for the old column names (var_name) and one for the actual numbers (value_name).

```python
import pandas as pd

# 1. Our starting "wide" data
df = pd.DataFrame({
    'Region': ['North', 'South'],
    'Apples': [15, 20],
    'Bananas': [15, 25]
})
```
