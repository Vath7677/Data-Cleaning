Project: "Messy E-Commerce Customer Database Rescue"
Data cleaning is the most critical phase. This project provides them with a deliberately broken, chaotic dataset that they must fix before it can be used for sales reports.
The Scenario: The company database crashed, and the backup data is a mess. You must clean the "Orders" dataset.
* Handling Missing Data: The dataset is riddled with NaNs. You must use .dropna() to remove rows completely missing an Order ID, .fillna() to replace missing shipping states and transaction with "Unknown" that were lost during export.
* Handling Duplicates: A glitch caused users to be charged twice. You must use .duplicated() to find identical transactions and .drop_duplicates(keep='first') to refund/remove the errors.
* Data Type Conversions: * "Price" is imported as a string (e.g., "$1,200.50"). You must remove the symbols and use pd.to_numeric() to turn it into a float.
    * Replace the “pending” and “unknown” with the previous datetime
    * Convert “purchase date" from messy text into a proper datetime object using pd.to_datetime().
    * Convert the “purchase date" format from various format into this format “M/D/Y”
* String Manipulation: The "Names" column has chaotic capitalization and extra spaces. You use .str.title() and .str.strip() to fix this. You must also extract the email domain (e.g., extracting "gmail.com" from "user@gmail.com") using .str.split().

```python
import pandas as pd
import numpy as np

# df = pd.read_csv('ecommerce.csv')
# df.to_excel('Cleaning.xlsx', index=False)

##          --->>>>>>>>  if u want to read an excel file u need to use .read_excel()

df = pd.read_excel('Cleaning.xlsx')

print(df.columns)
print(df.index)
df.shape
df.size
print(df.info())
df.head()
df.tail()
df.sample(5)

##          -->>>>>>>> the part of uisng  { .dropna() .any .all .subset }

dropna_all = df.dropna(how='all')      # The list do not have have all the columns  'NaNs'
dropna_all.head()
dropna_all

dropna_any = df.dropna(how='any')
dropna_any.head()
dropna_any                                   # after using dropna_any we see that the list have only { 675 item/ 1100 item }

# using .dropna and .fillna() on { id }

df_subnet = df.dropna(subset=['id'])          # id have only 930 after using .dropna(subset=['id'])
df_subnet

## --->>>>>>>  this part is talking about using .fillna() 

id_fillna = df.copy()             # i using this just to change name 
id_fillna['id'] = df['id'].fillna('unknow')
id_fillna

## --->>>>>>> the part that we using duplicated()

df[id_fillna.duplicated()]
id_fillna.sample(10)

##             --->>>>>>>>>   using to see duplicated in the row 
duplicated_item = df[id_fillna.duplicated(keep='first')]      # after using .duplicated { it is using to see an only the duplicated item only } 
duplicated_item.head()
duplicated_item.shape
duplicated_item

##  -->>>>>>>>>>>> after filtering the duplicated          #  .drop_duplicates() { we using to filtering the duplicated out }
data_clean = df.drop_duplicates(keep=False)
data_clean = df.drop_duplicates(keep='first')
data_clean

## -->>>>>>>> Data Type Conversions:

## ---->>>>>>>>>>. covert price to double and delete ('$')
df['price'] = df['price'].str.replace('$','').str.replace(',','')
df['price'] = pd.to_numeric(df['price'])

# ----->>>>>>> using to put NaT in to a null place 
df['purchase date'] = df['purchase date'].replace(['pending', 'unknow'], np.nan)
df['purchase date'] = df['purchase date'].ffill()


## ---->>>>>>>>>> Because our data have only 1100 and the null and worng standard is 910 ,so i decision to fill in by uisng {.ffill } 
df['purchase date'] = pd.to_datetime(df['purchase date'], errors='coerce')
df['purchase date'] = df['purchase date'].ffill()

##  --->>>>>>>>> using to set the ['purchase date' ] as month/day/year
df['purchase date'] = df['purchase date'].dt.strftime('%m/%d/%Y')
df.dropna(subset=['purchase date'])         # if null delete 
df

## ---->>>>>>>> using to count null or NaT 
nat_count = df['purchase date'].isnull().sum()    # we using this to count null
print(nat_count)

df.shape
df.sample(10)

# --->>>>>>>>>>>. String Manipulation: 

df['customer name'] = df['customer name'].str.strip().str.title()
# df

df[['User', 'Domain']] = df['email'].str.split('@', expand=True)
df.info()
df
print(df['User'])
print(df[['User', 'Domain']])

duplicated_item.to_excel('item_duplicated.xlsx', index=False)
```
